```python
def calculate_distance(X_train, X_test):
    dist_matrix = np.sqrt(np.sum((X_train[:, np.newaxis] - X_test) ** 2, axis=2))
    return dist_matrix

def predict_labels(dist_matrix, y_train, k):
    nearest_indices = np.argsort(dist_matrix, axis=0)[:k]
    nearest_labels = y_train[nearest_indices]
    y_pred = np.zeros(nearest_labels.shape[1])
    for i in range(nearest_labels.shape[1]):
        y_pred[i] = np.argmax(np.bincount(nearest_labels[:, i]))
    return y_pred.astype(int)

def knn_model_build(k, kf, X, y):
    accuracy_scores = []  
    confusion_matrices = []  

    for i, (train_index, test_index) in enumerate(kf.split(X)):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        dist_matrix = calculate_distance(X_train, X_test)
        y_pred = predict_labels(dist_matrix, y_train, k)
        accuracy = accuracy_score(y_test, y_pred)
        accuracy_scores.append(accuracy)

        confusion_matrices.append(confusion_matrix(y_test, y_pred))
        print(f"Confusion Matrix for Fold {i+1}:\n{confusion_matrices[-1]}")
        print(f"Accuracy for Fold {i+1}: {accuracy:.2f}\n")
    
    avg_accuracy = np.mean(accuracy_scores)
    print(f"Average Accuracy score of K Nearest Neighbors (K={k}): {avg_accuracy:.2f}")
   
    plt.figure(figsize=(10, 7))
    sns.heatmap(confusion_matrices[-1], cmap="Blues", annot=True, 
                xticklabels=["blues", "classical", "country", "disco", 
                             "hiphop", "jazz", "metal", "pop", "reggae", "rock"], 
                yticklabels=["blues", "classical", "country", "disco", 
                             "hiphop", "jazz", "metal", "pop", "reggae", "rock"])
    plt.title(f"Confusion Matrix for K Nearest Neighbors (K={k})", fontsize=16)
    plt.xlabel('Predicted Genre', fontsize=14)
    plt.ylabel('True Genre', fontsize=14)
    plt.show()

    return avg_accuracy, confusion_matrices
knn = KNeighborsClassifier(n_neighbors=10)
model_build(knn, split, 'K Nearest Neighbor')
```
